# **Task 2: Univariate Analysis**

# Distribution of age, watch_hours, monthly_fee, churned

```python
sns.set(style="whitegrid")
plt.figure(figsize=(16, 20))

plt.subplot(4, 2, 1)
sns.histplot(df['age'], kde=True, color='skyblue')
plt.title('Distribution of Age')

plt.subplot(4, 2, 2)
sns.histplot(df['watch_hours'], kde=True, color='orange')
plt.title('Distribution of Watch Hours')

plt.subplot(4, 2, 3)
sns.histplot(df['monthly_fee'], kde=True, color='green')
plt.title('Distribution of Monthly Fee')

plt.subplot(4, 2, 4)
sns.countplot(x='churned', data=df, palette='Set2')
plt.title('Churn Count (0 = No, 1 = Yes)')
plt.xticks([0, 1], ['No', 'Yes'])

plt.tight_layout()
plt.show()
```

# Count plots for categorical variables


```python
plt.figure(figsize=(16, 20))

plt.subplot(3, 2, 1)
sns.countplot(x='subscription_type', data=df, palette='muted')
plt.title('Subscription Type Count')

plt.subplot(3, 2, 2)
sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Gender Distribution')

plt.subplot(3, 2, 3)
sns.countplot(y='region', data=df, order=df['region'].value_counts().index, palette='cool')
plt.title('Region Distribution')

plt.subplot(3, 2, 4)
sns.countplot(x='device', data=df, palette='Set3')
plt.title('Device Usage')

plt.subplot(3, 2, 5)
sns.countplot(x='payment_method', data=df, palette='Accent')
plt.title('Payment Method Distribution')

plt.subplot(3, 2, 6)
sns.countplot(y='favorite_genre', data=df, order=df['favorite_genre'].value_counts().index, palette='Spectral')
plt.title('Favorite Genre Distribution')

plt.tight_layout()
plt.show()
```

# **Task 3: Bivariate Analysis**

# Compare Average watch_hours and monthly_fee

```python
plt.figure(figsize=(16, 18))
sns.set(style="whitegrid")

plt.subplot(3, 2, 1)
sns.barplot(x='subscription_type', y='watch_hours', data=df, estimator='mean', palette='Blues_d')
plt.title('Avg Watch Hours by Subscription Type')

plt.subplot(3, 2, 2)
sns.barplot(x='subscription_type', y='monthly_fee', data=df, estimator='mean', palette='Greens_d')
plt.title('Avg Monthly Fee by Subscription Type')

plt.subplot(3, 2, 3)
sns.barplot(x='region', y='watch_hours', data=df, estimator='mean', palette='Oranges_d')
plt.title('Avg Watch Hours by Region')
plt.xticks(rotation=45)

plt.subplot(3, 2, 4)
sns.barplot(x='region', y='monthly_fee', data=df, estimator='mean', palette='Purples_d')
plt.title('Avg Monthly Fee by Region')
plt.xticks(rotation=45)

plt.subplot(3, 2, 5)
sns.barplot(x='device', y='watch_hours', data=df, estimator='mean', palette='Reds_d')
plt.title('Avg Watch Hours by Device')


plt.subplot(3, 2, 6)
sns.barplot(x='device', y='monthly_fee', data=df, estimator='mean', palette='BuPu')
plt.title('Avg Monthly Fee by Device')

plt.tight_layout()
plt.show()
```

# Average avg_watch_time_per_day by favorite_genre

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='avg_watch_time_per_day', y='favorite_genre', data=df, estimator='mean', palette='coolwarm', ci=None)
plt.title('Average Daily Watch Time by Favorite Genre')
plt.xlabel('Avg Watch Time Per Day')
plt.ylabel('Favorite Genre')
plt.tight_layout()
plt.show()
```

# Churn Rate Analysis by Categorical Variables

```python
def churn_rate_by(col):
    return df.groupby(col)['churned'].mean().sort_values(ascending=False).reset_index()

churn_gender = churn_rate_by('gender')
churn_region = churn_rate_by('region')
churn_subscription = churn_rate_by('subscription_type')
churn_payment = churn_rate_by('payment_method')

fig, axs = plt.subplots(2, 2, figsize=(16, 12))

sns.barplot(x='gender', y='churned', data=churn_gender, ax=axs[0, 0], palette='Set2')
axs[0, 0].set_title('Churn Rate by Gender')

sns.barplot(x='region', y='churned', data=churn_region, ax=axs[0, 1], palette='Set3')
axs[0, 1].set_title('Churn Rate by Region')
axs[0, 1].tick_params(axis='x', rotation=45)

sns.barplot(x='subscription_type', y='churned', data=churn_subscription, ax=axs[1, 0], palette='Set1')
axs[1, 0].set_title('Churn Rate by Subscription Type')

sns.barplot(x='payment_method', y='churned', data=churn_payment, ax=axs[1, 1], palette='Pastel1')
axs[1, 1].set_title('Churn Rate by Payment Method')
axs[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

# **Task 4: Correlation Analysis**

# Create a heatmap of correlation between numerical variables.

```python
numerical_cols = ['age', 'watch_hours', 'last_login_days', 'monthly_fee',
                  'churned', 'number_of_profiles', 'avg_watch_time_per_day']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

# Look for potential predictors of churn, e.g., watch_hours, last_login_days, monthly_fee.


```python
churn_corr = correlation_matrix['churned'].drop('churned').sort_values(ascending=False)

print("🔍 Correlation of Numerical Features with 'churned':\n")
print(churn_corr)
```

# **Task 5: Insights & Recommendations**

# ***Key Insights from Netflix Customer Data Analysis***

**Users who watch less are more likely to leave:**

*Customers with lower total watch hours and shorter average daily watch time tend to churn more. It seems that less engaged users don’t find enough value in their subscription.*


**Inactive users are the biggest red flag:**

*One of the strongest predictors of churn is the number of days since the last login. The longer a user stays away from Netflix, the more likely they are to cancel.*


**Premium users are slightly more loyal than Basic users:**

*Surprisingly, Premium plan users have a slightly lower churn rate compared to Basic users. This could mean that users who invest more in their subscription are also more committed.*


**Africa has the most engaged audience:**

*On average, customers in Africa have the highest watch hours among all regions. This might reflect a strong streaming culture or fewer alternative entertainment options.*


**Mobile users tend to churn more than TV users:**

*Users who primarily stream on mobile devices show a higher churn rate compared to those watching on TVs. Larger screens might offer a better experience, encouraging retention.*


**Gift Card and Crypto users churn more:**

*Customers using non-traditional payment methods like gift cards or cryptocurrency are more likely to churn, possibly due to temporary subscriptions or less financial commitment.*


**Comedy and Drama lovers are more consistent:**

*People who favor Comedy and Drama genres tend to watch more frequently on a daily basis, suggesting strong content loyalty in these categories.*
